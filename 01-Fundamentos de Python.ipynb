## 1. ¿Qué es un lenguaje de programación?

Es una forma de **dar instrucciones a una computadora**. Utiliza palabras y reglas específicas para crear programas que le indican a la computadora qué hacer.

> **En palabras simples:** así como el español o el inglés sirven para que dos personas se comuniquen, un lenguaje de programación sirve para que un humano le hable a una máquina.

## 2. Sobre Python

### 2.1 Definición

**Python** es un lenguaje de programación de **alto nivel** (HLL), **interpretado**, **dinámicamente tipado** y **multiparadigma**, creado por **Guido van Rossum** en 1989 (publicado en 1991). Su filosofía está resumida en *"El Zen de Python"* (`import this`).

### 2.2 Características clave

| Característica | ¿Qué significa? |
|---|---|
| **Alto nivel** | Más cercano al lenguaje humano que al lenguaje máquina. |
| **Multiparadigma** | Soporta programación imperativa, orientada a objetos y funcional. |
| **Tipado dinámico** | Las variables no declaran su tipo (se infiere en tiempo de ejecución). |
| **Tipado fuerte** | No convierte tipos arriesgadamente (no puedes sumar `"5" + 3` sin convertir). |
| **Portátil** | Corre en Windows, macOS y Linux con el mismo código. |
| **Ecosistema enorme** | Más de 400.000 paquetes en PyPI. |

### 2.3 ¿Qué significa "interpretado"?

Tu archivo `.py` se compila a **bytecode** (`.pyc`), que es ejecutado por la **Python Virtual Machine (PVM)**. La implementación estándar se llama **CPython** (también existen PyPy, Jython e IronPython).

---


### 2.4 Lenguajes de alto nivel vs lenguajes de bajo nivel

> *Tabla equivalente a la imagen original; la información se mantiene de forma accesible y exportable.*

| Aspecto | Lenguaje de alto nivel (HLL) | Lenguaje de bajo nivel (LLL) |
|---|---|---|
| **Cercanía** | Cercano al lenguaje humano | Cercano al lenguaje máquina |
| **Sintaxis** | Natural y legible | Críptica, simbólica |

### 2.6 En Python, todo es un objeto

En Python, **absolutamente todo es un objeto**: los números, las cadenas, las listas, las funciones e incluso los módulos. Cada objeto tiene:

- Un **valor** (su contenido).
- Un **tipo** (la clase a la que pertenece, consultable con `type()`).
- Una **identidad** (su dirección en memoria, consultable con `id()`).
- **Atributos** y **métodos** accesibles por la *notación de punto* (`objeto.metodo()`).

Esto significa que incluso un número entero tiene métodos:


````python
# Hasta un número entero es un objeto con métodos
print((255).bit_length())     # cuántos bits necesita para representarse
print((10).to_bytes(2, "big")) # convertirlo a bytes
print(type(42))                # <class 'int'>
print(id(42))                  # dirección en memoria
````

### 6.1 Función `print()`

La función `print` ejecuta una **impresión en terminal** que nos ayuda a saber qué está pasando dentro del código y cómo va la ejecución.

Por simplicidad usaremos tres formas:

1. **Informativos**: puro texto.
2. **Variables**: imprimes el valor de una variable.
3. **Texto + variables (f-string)**: usando la letra `f` antes de las comillas.

> **Tip:** dentro del `print` también puedes ejecutar **operaciones aritméticas**, como `print(2 + 3)`.


````python
# Conversiones
entero = int("34")        # str -> int
decimal = float(56)        # int -> float
otro_decimal = float("95") # str -> float

# Tres formas de print
print("Esto es un entero")               # informativo
print(entero)                             # variable sola
print(f"Entero: {entero}, decimal: {decimal}, otro decimal: {otro_decimal}")  # f-string

# Conversión a cadena
cadena = str(entero)
print(f"el entero convertido a cadena es: {cadena} y es de tipo {type(cadena)}")

# Operaciones dentro del print
print(f"Suma: {2 + 3}, producto: {4 * 5}")
````

## 7. Operadores aritméticos

| Operación | Símbolo | Tipos que aceptan | Limitantes / Notas |
|---|---|---|---|
| **Suma** | `+` | `int`, `float`, `str` (concatena), `list` (concatena), `tuple` | `"5" + 3` → `TypeError`; tipos incompatibles fallan. |
| **Resta** | `-` | `int`, `float` | No aplica a `str` ni `list`. |
| **Multiplicación** | `*` | `int`, `float`, `str` (repite), `list` (repite) | `"hi" * 3` → `"hihihi"`; `str * float` → `TypeError`. |
| **División** | `/` | `int`, `float` | **Siempre devuelve `float`**. División por cero → `ZeroDivisionError`. |
| **División entera** | `//` | `int`, `float` | Devuelve la parte entera del cociente. División por cero → error. |
| **Módulo / Resto** | `%` | `int`, `float` | Devuelve el residuo. Útil para detectar pares/impares (`n % 2 == 0`). |
| **Exponenciación** | `**` | `int`, `float` | `2 ** 10 = 1024`. Cuidado con números muy grandes (overflow en `float`). |

### 7.1 Ejemplos


````python
a, b = 17, 5
print(f"Suma:           {a + b}")
print(f"Resta:          {a - b}")
print(f"Multiplicación: {a * b}")
print(f"División:       {a / b}")
print(f"División entera:{a // b}")
print(f"Módulo:         {a % b}")
print(f"Exponenciación: {a ** b}")

# Curiosidades:
print(f"\nstr * int:  {'ja' * 4}")
print(f"list + list: {[1,2] + [3,4]}")
````

## 8. Operadores de comparación / relacionales

Devuelven siempre un valor booleano (`True` o `False`).

| Operación | Símbolo | Tipos que aceptan | Limitantes / Notas |
|---|---|---|---|
| **Igual que** | `==` | Cualquier tipo comparable | Compara **valor**, no identidad. `[1,2] == [1,2]` → `True`. |
| **Diferente a** | `!=` | Cualquier tipo comparable | Inverso de `==`. |
| **Mayor que** | `>` | `int`, `float`, `str` (orden lexicográfico), `list` (elemento a elemento) | Mezclar `str` con `int` → `TypeError`. |
| **Menor que** | `<` | Igual que `>` | Igual restricción. |
| **Mayor o igual** | `>=` | Igual que `>` | – |
| **Menor o igual** | `<=` | Igual que `>` | – |
| **Identidad** | `is`, `is not` | Cualquier objeto | Compara si son el **mismo objeto**, no si valen lo mismo. |

### 8.1 Ejemplos


````python
a, b = 10, 7
print(f"a == b: {a == b}")
print(f"a != b: {a != b}")
print(f"a >  b: {a >  b}")
print(f"a <  b: {a <  b}")
print(f"a >= 10: {a >= 10}")
print(f"a <= 10: {a <= 10}")

# Comparación lexicográfica de strings (orden alfabético)
print(f"'manzana' < 'pera': {'manzana' < 'pera'}")

# is vs ==
x = [1, 2, 3]
y = [1, 2, 3]
print(f"x == y: {x == y}")    # True (mismo valor)
print(f"x is y: {x is y}")    # False (objetos distintos)
````

## 9. Operadores lógicos

Combinan o niegan valores booleanos. Son la base para construir condiciones complejas.

| Operación | Símbolo | Descripción | Ejemplo |
|---|---|---|---|
| **Y (conjunción)** | `and` | `True` solo si **ambas** son verdaderas. | `True and False` → `False` |
| **O (disyunción)** | `or` | `True` si **al menos una** es verdadera. | `True or False` → `True` |
| **Negación** | `not` | Invierte el valor booleano. | `not True` → `False` |

> **Tip:** Python usa **evaluación de cortocircuito** — si la primera parte de un `and` es `False`, no evalúa la segunda; lo mismo con `or` si la primera es `True`.

### 9.1 Ejemplos


````python
edad = 25
tiene_id = True

print(edad >= 18 and tiene_id)         # True
print(edad >= 18 or  tiene_id)         # True
print(not tiene_id)                    # False

# Combinaciones
puede_comprar = (edad >= 18) and tiene_id and (edad < 65)
print(f"¿Puede comprar?: {puede_comprar}")

# Tabla de verdad
print("\nTabla de verdad de AND:")
for a in (True, False):
    for b in (True, False):
        print(f"  {a} and {b} = {a and b}")
````

## 10. Condicional `if / elif / else`

Los **condicionales** nos ayudan a ejecutar ciertas líneas de código cuando se cumple una condición. Así, el programa toma decisiones dependiendo de los valores con los que trabaja.

### 10.1 Sintaxis básica

```python
if condicion:
    # bloque que se ejecuta si la condición es True
    instrucciones

elif otra_condicion:
    # bloque alternativo (puede haber varios elif)
    instrucciones

else:
    # bloque que se ejecuta si NINGUNA condición anterior se cumplió
    instrucciones
```

> **Reglas importantes:**
> 1. La línea de `if`, `elif` y `else` **siempre termina con dos puntos `:`**.
> 2. El bloque debajo se identifica por **indentación** (4 espacios).
> 3. `elif` y `else` son **opcionales**.

### 10.2 Ejemplo


````python
edad = 20

if edad < 13:
    categoria = "niño"
elif edad < 18:
    categoria = "adolescente"
elif edad < 65:
    categoria = "adulto"
else:
    categoria = "adulto mayor"

print(f"Con {edad} años, eres: {categoria}")

# Forma corta (operador ternario)
estado = "mayor" if edad >= 18 else "menor"
print(f"Estado: {estado}")
````

La cadena de documentación de Python, comúnmente conocida como docstring, es un literal de cadena, y se utiliza en la definición de clases, módulos, funciones o métodos. Las docstrings son accesibles desde el atributo doc (__doc__) para cualquiera de los objetos Python y también con la función incorporada help().

````python
# Esto es un comentario de una línea

def saludar(nombre):
    """Función que devuelve un saludo personalizado.

    Este texto entre triples comillas es un docstring.
    """
    return f"Hola, {nombre}"

print(saludar("Samael"))
print(saludar.__doc__)        # acceso al docstring
````

````python
print(help(saludar))
````
