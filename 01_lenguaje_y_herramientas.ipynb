## ¿Que es un lenguaje de programación?

Es una forma de dar instrucciones a una computadora. Utiliza palabras y reglas específicas para crear programas que le indican a la computadora qué hacer.

## 1. Definición

Python es un lenguaje de programación de **alto nivel** (HLL), **interpretado**, **dinámicamente tipado** y **multiparadigma**, creado por Guido van Rossum en 1989 (publicado en 1991). Su filosofía está resumida en *"El Zen de Python"* (`import this`).

**Características clave:**

- Alto nivel: mas asimilado al lenguaje humano.
- Multiparadigma (imperativo, orientado a objetos, funcional).
- Tipado dinámico (las variables no declaran tipo) pero fuerte (no convierte tipos arriesgadamente).
- Portátil: corre en Windows, macOS, Linux.
- Ecosistema enorme (>400.000 paquetes en PyPI).

**¿Qué significa "interpretado"?** Tu archivo `.py` se compila a *bytecode* (`.pyc`) que ejecuta la **Python Virtual Machine (PVM)**. La implementación estándar se llama **CPython** (existen también PyPy, Jython, IronPython).

# Sintaxis de Python

Reglas de escritura del lenguaje y convenciones de estilo.

## Definición

Python tiene reglas sintácticas estrictas pero pocas. Las más importantes:

1. **Sensibilidad a mayúsculas (case-sensitive):** `edad`, `Edad` y `EDAD` son tres variables distintas.
2. **Notación de punto (dot notation):** `objeto.atributo` lee algo del objeto, `objeto.metodo()` ejecuta una acción.
3. **Palabras reservadas (keywords):** no puedes usarlas como nombres. Lista completa con `keyword.kwlist`.
4. **Indentación obligatoria:** los bloques se delimitan por sangría (4 espacios estándar), no con llaves `{}`.
5. **Comentarios:** `#` para una línea, triple comilla `""" ... """` para bloques o docstrings.
6. **Convenciones de nombres (PEP 8):** `snake_case` para variables y funciones, `PascalCase` para clases, `MAYUSCULAS` para constantes.

### 1.- Sensibilidad a mayúsculas

### 4.- Indentación correcta

```python
edad = 20
if edad >= 18:
    print("Mayor de edad")    # 4 espacios → pertenece al if
    print("Puede votar")      # mismo nivel → mismo bloque
print("Fin del programa")     # sin indentar → fuera del if
```

### 5.- Comentarios y docstrings

La cadena de documentación de Python, comúnmente conocida como docstring, es un literal de cadena, y se utiliza en la definición de clases, módulos, funciones o métodos. Las docstrings son accesibles desde el atributo doc (__doc__) para cualquiera de los objetos Python y también con la función incorporada help().

```python
# Esto es un comentario de una línea

def saludar(nombre):
    """Función que devuelve un saludo personalizado.

    Este texto entre triples comillas es un docstring.
    """
    return f"Hola, {nombre}"

print(saludar("Samael"))
print(saludar.__doc__)        # acceso al docstring
```

```python
print(help(saludar))
```

## 3. Consideraciones importantes

- **Python 2 está descontinuado** desde 2020. Usa siempre Python 3 (3.10+ idealmente).
- **CPython** es la referencia; cuando alguien dice "Python", asume CPython salvo aclaración.
- En Windows, si `python` no funciona en la terminal, reinstala marcando *"Add Python to PATH"*.
- En macOS/Linux ya viene un Python del sistema; **no instales librerías ahí**, usa entornos virtuales.
- Jupyter y Colab son ideales para aprender y para ciencia de datos; VS Code es ideal para proyectos.
- `pip` es el gestor de paquetes oficial; `conda` es la alternativa cuando usas Anaconda.

### Errores típicos en esta etapa

- `python: command not found` o `'python' no se reconoce`: problema de PATH.
- `ModuleNotFoundError`: la librería no está instalada en el entorno activo.
- Conflicto entre versiones: usa `python3` y `pip3` cuando coexistan Python 2 y 3.
- Kernel de Jupyter apuntando al Python equivocado: revisa con `!which python` (o `!where python` en Windows).
